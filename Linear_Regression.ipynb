```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Read data
data = pd.read_csv('incomedata.csv')
print(data.shape)
data.head()
```

```python
X = data['income'].values
Y = data['Easeofliving'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)
m=data.shape[0]

num=0
den=0

for i in range (m):
    num += (X[i]-mean_x)*(Y[i]-mean_y)
    den += (X[i]-mean_x)**2

b1=num/den
b0=mean_y-(b1*mean_x)

print(b1,b0)
```

```python
plt.rcParams['figure.figsize'] = (10.0,10.0)
max_x = np.max(X) + 5
min_x = np.min(X) - 5

x = np.linspace(min_x,max_x,1000)
y=b0 + b1*x

plt.plot(x,y,color="#58b970",label="Regression Line")

plt.scatter(X,Y,color="#ef5423",label="Scatter Plot")

plt.xlabel("Income")
plt.ylabel("Ease of Living")
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range (m):
    y_pred = b0 + b1 * X[i]
    ss_t += (Y[i]-mean_y)**2
    ss_r += (Y[i]-y_pred)**2
r2 = 1 - (ss_r/ss_t)
print(r2)
```
